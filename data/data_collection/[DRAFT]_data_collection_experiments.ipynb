```python
import requests
import json
import csv

alpha_vantage_api_key = 'your_alpha_vantage_api_key'
stocks_to_track = ['MSFT',]  

url = 'https://www.alphavantage.co/query'
params = {
    'apikey': alpha_vantage_api_key,
    'function': 'NEWS_SENTIMENT',
    'tickers': ','.join(stocks_to_track),
    'limit': "100",
    'time_from': '20220601'
}

response = requests.get(url, params=params)

if response.status_code == 200:
    data = response.json()
    news_articles = data.get('data', []) 

    if news_articles:
        csv_file_name = 'alpha_vantage_news_articles.csv'

        with open(csv_file_name, 'w', newline='', encoding='utf-8') as csv_file:
            csv_writer = csv.DictWriter(csv_file, fieldnames=['ticker', 'title', 'date', 'source', 'url'])
            
            csv_writer.writeheader()

            for article in news_articles:
                csv_writer.writerow({
                    'ticker': article.get('symbol'),
                    'title': article.get('title'),
                    'date': article.get('date'),
                    'source': article.get('source_name'), 
                    'url': article.get('news_url'),
                })

        print(f"Alpha Vantage news articles have been successfully saved to {csv_file_name}")
    else:
        print(f"No news articles found in the response.")
else:
    print(f"Error: {response.status_code} - {response.text}")
```

```python
import requests
import csv

api_token = '655add3de97bb4.41000237' 
ticker_to_track = 'AAPL.US' 

url = f'https://eodhd.com/api/news'

params = {
    'api_token': api_token,
    's': ticker_to_track,
    'limit': "1000",
    'fmt': 'json',
    'offset': '0'
}

response = requests.get(url, params=params)


if response.status_code == 200:
    data = response.json()
    print(data)

    news_articles = data if isinstance(data, list) else data.get('news', [])

    if news_articles:
        csv_file_name = 'financial_news_articles.csv'

        with open(csv_file_name, 'w', newline='', encoding='utf-8') as csv_file:
            csv_writer = csv.DictWriter(csv_file, fieldnames=['date', 'title', 'content', 'link', 'symbols'])
            
            csv_writer.writeheader()

            for article in news_articles:
                csv_writer.writerow({
                    'date': article.get('date'),
                    'title': article.get('title'),
                    'content': article.get('content'),
                    'link': article.get('link'),
                    'symbols': ','.join(article.get('symbols', [])),
                })

        print(f"Financial news articles have been successfully saved to {csv_file_name}")
    else:
        print(f"No news articles found in the response.")
else:
    print(f"Error: {response.status_code} - {response.text}")
```

```python
import requests
import csv
import os

api_token = '655add3de97bb4.41000237'

tickers = ['AAPL', 'GOOGL', 'MSFT', 'AMZN', 'TSLA', 'SPY', 'QQQ', 'JPM', 'BAC', 'GS', 'C']

csv_file_name = 'financial_news_data_eodhd_all.csv'

with open(csv_file_name, 'w', newline='', encoding='utf-8') as csv_file_all:
    csv_writer_all = csv.writer(csv_file_all)
    csv_writer_all.writerow(['ticker', 'date', 'title', 'content', 'link', 'symbols'])

for ticker in tickers:
    url = f'https://eodhistoricaldata.com/api/news'
    params = {
        "api_token": api_token,
        "symbol": ticker,
        "from": "2022-01-01",
        "to": "2022-06-30",
        'limit': '1000'
    }

    response = requests.get(url, params=params)


    if response.status_code == 200:
        data = response.json()

        with open(csv_file_name, 'a', newline='', encoding='utf-8') as csv_file_all:
            csv_writer_all = csv.writer(csv_file_all)
            for news_item in data:
                if 'date' in news_item and 'title' in news_item:
                    csv_writer_all.writerow([ticker, news_item.get('date', ''), news_item.get('title', ''), news_item.get('content', ''), news_item.get('link', ''), news_item.get('symbols', '')])
                else:
                    print(f"Missing required keys for {ticker}: {news_item}")

        ticker_csv_file_name = f'financial_news_data_eodhd_{ticker}.csv'
        with open(ticker_csv_file_name, 'w', newline='', encoding='utf-8') as csv_file_ticker:
            csv_writer_ticker = csv.writer(csv_file_ticker)
            csv_writer_ticker.writerow(['date', 'title', 'content', 'link', 'symbols'])
            for news_item in data:
                if 'date' in news_item and 'title' in news_item:
                    csv_writer_ticker.writerow([news_item.get('date', ''), news_item.get('title', ''), news_item.get('content', ''), news_item.get('link', ''), news_item.get('symbols', '')])
                else:
                    print(f"Missing required keys for {ticker}: {news_item}")
    else:
        print(f"Error for {ticker}: {response.status_code} - {response.text}")

print(f"Financial news data from EODHD API has been successfully saved to {csv_file_name} and individual files for each ticker.")
```
